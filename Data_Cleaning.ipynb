```python
import pandas as pd
```

```python
df = pd.read_excel("/content/Customer_Call_List.xlsx")
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns="Not_Useful_Column")
df
```

```python
# remove errors from Last_Name
df["Last_Name"] = df["Last_Name"].str.strip("./_")
df
```

```python
# remove errors from Phone_Number
# df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '')
# df
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex=True)
df
```

```python
# Convert all elements in 'Phone_Number' column to strings
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))

# Formatting Phone_Number
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
# replace nan-- or Na-- values to nothing in Phone_Number

df["Phone_Number"] = df["Phone_Number"].str.replace('(nan--)|(Na--)', '', regex=True)
df
```

```python
# Rearrange and split address
df[["Street_Name", "State", "Zip_Code"]] = df["Address"].str.split(',', n=2, expand=True)
# Drop old Address Column
df = df.drop(columns="Address")
df
```

```python
# Change Yes and No to Y and N in the columns (Paying Customer and Do Not Contact)
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes|No', lambda x: 'Y' if x == 'Yes' else 'N', regex=True)
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes|No', lambda x: 'Y' if x == 'Yes' else 'N', regex=True)

# Fix null values
df = df.replace('N/a', '')
df = df.fillna('')
df
```

```python
# Drop all Cutomers without a Phone_Number
for x in df.index:
  if df.loc[x, "Phone_Number"] == '':
    df.drop(x, inplace=True)

df
```

```python
# Reset index numbers and replace with current one
df = df.reset_index(drop=True)
df
```
